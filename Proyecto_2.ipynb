### Ejercicio No 3

- Crear una función que permita generar 3 variables cualitativas ordinales a partir de las variables numéricas con los siguientes criterios:

    ★ Si es mayor a la “media + 1*desviación estándar” → Alto

    ★ Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio

    ★ Si es menor a la “media - 1*desviación estándar” → Bajo
    
Nota: Mostrar con las nuevas variables generadas.

Elegir con libertad las variables a trabajar.

La función debe contar con su respectiva documentación.

```python
import pandas as pd
"""Importamos las librerías que necesitamos"""
```

```python
#las variables cualitativas ordinales, ejemplo: "Excelente, bueno, regular, malo", o sea, hay que asignar alto, medio o bajo dependiendo de los criterios datos
def var_ordinales(df):

    # Calcular la media y desviación estándar para cada variable numérica
    media = df.mean()
    desviacion_estandar = df.std()

    # Generar nuevas variables cualitativas ordinales
    for variable in df.columns: #Itera en todas las columnas del DataFrame
       #verifica si la variable es una variable numérica de las que se especifican
        if variable in ['Millas_Por_Galon', 'Cilindros', 'Peso_En_Libras', 'Desplazamiento', 'Aceleracion', 'Caballo_De_Fuerza']:
          #la f es para poder utilizar f-strings, las cuales permiten tomar un valor y agregarlo a un texto, como se muestra a continuación
            nueva_var = f"{variable}_ordinal" #Nombre para la nueva columna, en donde añade ordinal
            df[nueva_var] = pd.cut(df[variable], #Se crea nueva columna
            #pd.cut nos ayuda a asignar los valores ordinales que en este caso son 'Bajo', 'Medio', 'Alto'
                                        bins=[-float("inf"), media[variable] -
                                              desviacion_estandar[variable], media[variable] +
                                              desviacion_estandar[variable], float("inf")],
                                        #Intervalo de los datos, de -infinito hasta infinito
                                        labels=['Bajo', 'Medio', 'Alto'], #Variables Ordinarias
                                        include_lowest=True)

    return df #nos devuelve el dataframe nuevo, con las operaciones que le aplicamos

"""Para crear las tres variables cualitativas ordinales se hace lo siguiente:
      - Se define la función "var_ordinales", con el argumento de "df"
      - Se calcula la media y desviación estándar de cada variable numérica
      - Iteramos todas las columnas hasta encontrar todas las que se definieron como cuantitativas
      - Se asigna un nombre nuevo, respecto al nombre pasado
      - Se crea la columna para asignar las nuevas variables dependiendo si la operación da Bajo, Medio o Alto.
      - Se devuelve un Dataframe nuevo respecto a los datos"""

# Cargar el DataFrame desde el archivo CSV
dfAutos2 = pd.read_csv('autos v2.csv')

# Aplica la función que crea a las variables ordinales
dfAutos2 = var_ordinales(dfAutos2)

# DataFrame final
print(dfAutos2)

```
